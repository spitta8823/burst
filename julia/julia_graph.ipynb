```julia
function pprocess_poisson(lambda)
    xs = [0.0];

    L = 1;
    dt = 100
    x = 0.0;
    while x <= L*dt
        x = x + - log(rand()) / lambda;
        push!(xs, x);
    end
    
#    return xs
    i = 1;
    τ_a = [];
    while i <= length(xs) -1
        τ = xs[i+1] - xs[i]
        push!(τ_a, τ); 
        i = i + 1
    end
    
    return sort!(τ_a)
end
```

```julia
τ = pprocess_poisson(10000)
```
